7\. Click this link https://www.who.int/news-room/releases/1, which will lead you to the News Releases page of the World Health Organization (WHO). On that page, you should see ten links to the most recent news release pages. Click the page number 2 at the bottom, which will lead you to the second page, and you will see another ten links, and so on and so forth. Create a CSV file named hw2_output.csv in the outcome folder, such that the header row contains the two column names, title and url, and the following 50 rows contain the titles and full URLs of all the links in pages 1 to 5. For the CSV file, columns are separated by a tab and rows by a new line. While fetching the content from the web site, sleep for one second per page. The title text may have unnecessary strings like '\r' and '\n', so remove them all in the title text.   

For example, the first value in the title column is 'WHO launches “One life, one liver” campaign on World Hepatitis Day' and the first value in the url column is 'https://www.who.int/news/item/28-07-2023-who-launches--one-life--one-liver--campaign-on-world-hepatitis-day'. Note that the titles and URLs could be different than yours, as the content of the webpage changes overt time.

```python
# Your answer here

# hw2_output.csv 파일에 page 1에서 5까지의 title과 url을 저장
# column: title, url
# 열의 separator는 \t, row의 separator는 \n
# 한 페이지 당 time.sleep(1)
# title의 \r, \n는 아예 제거

import time

with open(f"{outcome_folder}/hw2_output.csv", "w") as fw: # 맨 윗줄 먼저 써주기
    fw.write("title\turl\n")

for i in range(1, 6): # 5개의 page

  r = requests.get("https://www.who.int/news-room/releases/" + str(i)) # i번째 page의 url로 요청
  soup = BeautifulSoup(r.content, "html.parser")

  total_title = soup.find('div', {"class": "list-view vertical-list vertical-list--image"}).find_all('p')
  titles = [] #10개의 title
  for title in total_title:
    titles.append(title.text.replace('\r', '').replace('\n', '').strip())

  total_url = soup.find('div', {'class': 'list-view vertical-list vertical-list--image'}).find_all('a')
  urls = [] #10개의 url
  for url in total_url:
    url = "https://www.who.int/" + url['href'] # 앞 부분 추가해줘야함
    urls.append(url)

  outcome_folder = f"{path}/{my_folder}/outcome"
  with open(f"{outcome_folder}/hw2_output.csv", "a") as fw: # append mode로 open

    for i in range(10):
      title = titles[i]; url = urls[i]
      fw.write(f"{title}\t{url}\n")

  time.sleep(1)
```
